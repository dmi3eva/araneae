```python
def check(filename: str):
    en_spider = EnSpiderDB()
    ru_spider = RuSpiderDB()
    with open(filename, 'r', encoding='utf-8') as f:
        data = json.load(f)
    for ind, sample in enumerate(data):
        sql_en = sample['query']['en']
        sql_ru = sample['query']['ru']
        # print(sql_en)
        if 'count' in sql_ru.lower():
            print(f"{sample['id']}")
            print(f"{sample['db_id']}")

            print(f"{sql_en}")
            en_result = en_spider.execute_request(sample["db_id"], sql_en)
            print(en_result)
            print(f"{sql_ru}")
            ru_result = ru_spider.execute_request(sample["db_id"], sql_ru)
            print(ru_result)  
            print()
```
